# Machine-Learning Analysis

This notebook compares raw and dentin-normalized radiomic features using nested stratified CV. The manuscript-matched feature screening step uses Levene's test to choose Student versus Welch two-sample t-test and retains features with p < 0.05. Feature screening, scaling, SMOTE, and hyperparameter tuning are all placed inside the training folds. No missing-value imputation is applied.

```python
from pathlib import Path

from src.ml_analysis import run_all_experiments
```

```python
RAW_FEATURE_CSV = Path('data/features/raw_radiomic_features.csv')
NORMALIZED_FEATURE_CSV = Path('data/features/dentin_normalized_radiomic_features.csv')
OUTPUT_DIR = Path('results/ml_analysis')

outputs = run_all_experiments(
    raw_csv=RAW_FEATURE_CSV,
    normalized_csv=NORMALIZED_FEATURE_CSV,
    output_dir=OUTPUT_DIR,
    label_column='diagnosis',
    cyst_suffix='C',  # fallback only if diagnosis is not included in feature CSV
    model_names=['logistic_regression', 'xgboost'],
    random_state=2048,
    use_smote=True,
)

outputs['pooled_metrics'].sort_values(['feature_set', 'model', 'dataset'])
```
